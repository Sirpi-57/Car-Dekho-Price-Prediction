### Check How many data are outside the IQR Boundary:

```python
import pandas as pd

def count_outliers_iqr(df):
    outlier_counts = {}
    
    for column in df.select_dtypes(include=['float64', 'int64']).columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Count outliers
        outlier_count = ((df[column] < lower_bound) | (df[column] > upper_bound)).sum()
        outlier_counts[column] = outlier_count
    
    return outlier_counts

# Example usage:
outlier_counts = count_outliers_iqr(df)
print(outlier_counts)
```

### Skewness of the Data:

```python
numerical_columns = ['km',
                     'Mileage']

# Plotting distribution and skewness for each numerical variable
plt.figure(figsize=(14, 10))

for i, col in enumerate(numerical_columns, 1):
    plt.subplot(4, 3, i)
    sns.histplot(df[col], kde=True, color='blue')
    plt.title(f'Distribution of {col}')
    plt.xlabel('')
    plt.tight_layout()

plt.suptitle('Distribution and Skewness of Numerical Variables', y=1.02)
plt.show()
```

### The distributions in the graphs show a high level of skewness, especially in variables like km, price and Mileage. 

### Owner no. , Year of Registration, Manufacture, Seats, are not numerical in nature but categorical. hence lets not consider them now.

### 1. Log Transformation:
#### For highly skewed data (such as km, Displacement), applying a log transformation can reduce skewness and normalize the distribution. This works best for variables that do not contain zero or negative values.

### 2. Square Root Transformation:
#### Another method to reduce skewness is a square root transformation, which can be applied to moderately skewed data. It is less aggressive than log transformation.

### Since Price is Target variable no need to apply for it now.

```python
# Step 2: Ordinal Encoding for 'Insurance Validity'
insurance_order = {
    'Comprehensive': 3,
    'Third Party': 2,
    'Zero Dep': 1,
    'Not Available': 0
}
df['Insurance Validity'] = df['Insurance Validity'].map(insurance_order)

# Step 3: Ordinal Encoding for 'ownerNo' (reversed)
owner_no_order = {1: 5, 2: 4, 3: 3, 4: 2, 5: 1}
df['ownerNo'] = df['ownerNo'].map(owner_no_order)

# Step 4: Ordinal Encoding for 'Seats_1'
# Since there's no clear relationship between seats and price, you can treat it as a categorical variable.
df['Seats_1'] = df['Seats_1'].astype('category')

# Step 5: Updated Ordinal Encoding for 'Fuel Type'
fuel_order = {
    'Diesel': 5,
    'Petrol': 4,
    'CNG': 3,
    'Electric': 2,
    'LPG': 1
}
df['Fuel Type'] = df['Fuel Type'].map(fuel_order)

# Step 6: One-Hot Encoding for 'bt', 'transmission', 'oem', 'state'
df = pd.get_dummies(df, columns=['bt', 'transmission', 'oem', 'state'], drop_first=True)

# Display the first few rows of the updated DataFrame
print(df.head())
```
