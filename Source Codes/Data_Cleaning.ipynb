##### 1. print the null rows where Max Power and Torque  is null. 
##### 2. find what is the oem and model and Year of Manufacture, when Max Power and Torque is null.
##### 3. check if there is other rows where the same oem,model, Year of Manufacture has Max Power and Torque and print them.

```python
# Step 1: Print the null rows where Max Power and Torque are null
null_power_torque = df[df['Max Power'].isnull() & df['Torque'].isnull()]

print("Rows where Max Power and Torque are null:")
print(null_power_torque)
```

```python
# Step 2: Find OEM, Model, and Year of Manufacture when Max Power and Torque are null
oem_model_year_details = null_power_torque[['oem', 'model', 'Year of Manufacture']].drop_duplicates()

print("\nOEM, Model, and Year of Manufacture for rows with null Max Power and Torque:")
print(oem_model_year_details)
```

```python
# Step 3: Check for other rows with the same OEM, Model, Year of Manufacture
matching_rows = []  # To store matching rows after updates

for index, row in oem_model_year_details.iterrows():
    oem = row['oem']
    model = row['model']
    year_of_manufacture = row['Year of Manufacture']
    
    similar_rows = df[
        (df['oem'] == oem) &
        (df['model'] == model) &
        (df['Year of Manufacture'] == year_of_manufacture) &
        (df['Max Power'].notnull()) &
        (df['Torque'].notnull())
    ]

    if not similar_rows.empty:
        print(f"\nSimilar rows for OEM: {oem}, Model: {model}, Year of Manufacture: {year_of_manufacture}:")
        print(similar_rows)
        matching_rows.append(similar_rows)  # Store matching rows
    else:
        print(f"\nNo similar rows found for OEM: {oem}, Model: {model}, Year of Manufacture: {year_of_manufacture} with non-null Max Power and Torque.")

```

```python
# Step 4: Fill null values and remove rows with both nulls
null_filled_count = 0  # Count of null values filled
deleted_rows_count = 0  # Count of rows deleted

# Iterate over each entry with null values in Max Power and Torque
for index, row in oem_model_year_details.iterrows():
    oem = row['oem']
    model = row['model']
    year_of_manufacture = row['Year of Manufacture']
    
    # Find similar rows with non-null Max Power and Torque
    similar_rows = df[
        (df['oem'] == oem) &
        (df['model'] == model) &
        (df['Year of Manufacture'] == year_of_manufacture) &
        (df['Max Power'].notnull()) &
        (df['Torque'].notnull())
    ]

    if not similar_rows.empty:
        print(f"\nFilling null values for OEM: {oem}, Model: {model}, Year of Manufacture: {year_of_manufacture}")
        
        # Fill the null values for each column
        for col in ['Max Power', 'Torque']:
            nulls_before = df[col].isnull().sum()  # Null count before filling
            
            # Fill nulls
            df.loc[
                (df['oem'] == oem) & 
                (df['model'] == model) & 
                (df['Year of Manufacture'] == year_of_manufacture) & 
                (df[col].isnull()), 
                col
            ] = similar_rows[col].values[0]  # Use the first similar value
            
            nulls_after = df[col].isnull().sum()  # Null count after filling
            filled = nulls_before - nulls_after  # Number of nulls filled
            
            # Print for debugging
            print(f"Filled {filled} nulls in column {col}")
            null_filled_count += filled  # Update count of filled values

    else:
        # Count rows to be deleted
        rows_to_delete = df[
            (df['oem'] == oem) & 
            (df['model'] == model) & 
            (df['Year of Manufacture'] == year_of_manufacture) & 
            (df['Max Power'].isnull() & df['Torque'].isnull())
        ].shape[0]
        
        print(f"\nDeleting {rows_to_delete} rows for OEM: {oem}, Model: {model}, Year of Manufacture: {year_of_manufacture} with null Max Power and Torque.")
        
        # Remove rows
        df = df[~((df['oem'] == oem) & (df['model'] == model) & (df['Year of Manufacture'] == year_of_manufacture) & (df['Max Power'].isnull() & df['Torque'].isnull()))]
        
        deleted_rows_count += rows_to_delete  # Update deleted row count

# Print final counts
print(f"\nTotal null values filled in Max Power and Torque: {null_filled_count}")
print(f"Total rows deleted (both Max Power and Torque were null): {deleted_rows_count}")

# Optional: Show DataFrame after filling/removal for verification
#print(df)

```

```python
import re

# Function to clean and convert price values to integer
def clean_price(price_str):
    # Remove rupee symbol (₹) and commas
    price_str = price_str.replace('₹', '').replace(',', '').strip()

    # Handle Lakh and Crore cases
    if 'Lakh' in price_str:
        price_str = price_str.replace('Lakh', '').strip()  # Remove 'Lakh'
        return int(float(price_str) * 100000)  # Convert Lakh to the respective value
    elif 'Crore' in price_str:
        price_str = price_str.replace('Crore', '').strip()  # Remove 'Crore'
        return int(float(price_str) * 10000000)  # Convert Crore to the respective value
    else:
        # Handle plain numbers (already in thousands, no Lakh/Crore)
        return int(price_str)

# Apply the clean_price function to the 'price' column
df['price'] = df['price'].apply(clean_price)

# Display the cleaned DataFrame
print(df)
```

### 16.Max Power:  

```python
# Function to clean and convert max power
def convert_power(value):
    # Normalize the string
    value = value.replace(' at ', ' ').replace(' rpm', '').replace('rpm', '').strip()
    # Use regex to remove anything after bhp, PS, HP, or parentheses
    bhp_match = re.search(r'(\d*\.?\d+)\s*bhp', value)
    ps_match = re.search(r'(\d*\.?\d+)\s*PS', value)
    hp_match = re.search(r'(\d*\.?\d+)\s*HP', value)
    parenthesis_match = re.search(r'\((\d*\.?\d+)\)', value)
    
    if bhp_match:
        return round(float(bhp_match.group(1)), 2)
    elif ps_match:
        return round(float(ps_match.group(1)) * 0.98632, 2)
    elif parenthesis_match:
        return round(float(parenthesis_match.group(1)), 2)
    elif hp_match:
        return round(float(hp_match.group(1)) * 0.98632, 2)
    else:
        # Check for numerical values or other formats
        numeric_match = re.search(r'(\d*\.?\d+)', value)
        if numeric_match:
            return round(float(numeric_match.group(1)), 2)
    
    return None  # Return None for unhandled formats

# Apply the conversion function to the 'Max Power' column
df['Max Power'] = df['Max Power'].apply(convert_power)

# Verify the change
print(df['Max Power'].dtype)  # Should show float64

# Display the updated DataFrame
print(df)
```

```python
#String Normalization:

#Replaces instances of at and rpm with spaces, then removes any trailing spaces.
#Regex Extraction:

#Searches for patterns in the string using regular expressions:
#For bhp, it captures the numeric value followed by bhp.
#For PS, it captures the numeric value followed by PS and converts it.
#For HP, it captures the numeric value followed by HP and converts it.
#It also handles values within parentheses to extract the numeric value.
#Handling Ranges:

#If the value includes a range, only the first numeric value is used.
#Return Value:

#Returns None for unhandled formats.
```

### 17.Torque:  

### Error Types:

##### Here’s a one-line summary for each type of conversion logic in the Torque column:

##### 172nm: Extract the numeric value and convert it to float.

##### 99.04Nm: Extract the numeric value before "Nm" and convert it to float.

##### 33: Convert directly to float.

##### 146Nm at 4800 rpm: Extract the numeric value before "Nm" and convert it to float.

##### 259.87NM: Extract the numeric value before "NM" and convert it to float.

##### 22.4 kgm at 1750-2750rpm: Convert the numeric value from "kgm" to "Nm" using ×9.80665.

##### 11.4 kgm at 4,000 rpm: Convert the numeric value from "kgm" to "Nm" using ×9.80665.

##### 25 KGM: Convert the numeric value from "KGM" to "Nm" using ×9.80665.

##### 155 Nm at 1600-2800 rpm: Extract the numeric value before "Nm" and convert it to float.

##### 210 / 1900: Extract "210" and convert it to float.

##### 110 (11.2): Extract "110" and convert it to float.

##### 35.7: Convert directly to float.

##### 14.9 KGM at 3000 RPM: Convert the numeric value from "KGM" to "Nm" using ×9.80665.


```python
# Function to clean and convert torque
def convert_torque(value):
    # Normalize the string
    value = value.replace(' at ', ' ').strip()
    
    # Remove unwanted parts like 'rpm'
    value = re.sub(r'\s*rpm.*$', '', value)  # Remove everything after 'rpm'
    value = re.sub(r'\s+', ' ', value)  # Normalize spaces

    # Handle kgm conversion to Nm
    kgm_match = re.search(r'(\d*\.?\d+)\s*(?:kgm|KGM|Kgm)', value)
    nm_match = re.search(r'(\d*\.?\d+)\s*(?:Nm|nm|NM)', value)
    numeric_match = re.search(r'(\d*\.?\d+)', value)  # For numeric values without units

    if nm_match:
        return round(float(nm_match.group(1)), 2)  # Return value in Nm
    elif kgm_match:
        return round(float(kgm_match.group(1)) * 9.80665, 2)  # Convert kgm to Nm
    elif numeric_match:
        return round(float(numeric_match.group(1)), 2)  # Return standalone numeric value

    return None  # Return None for unhandled formats

# Apply the conversion function to the 'Torque' column
df['Torque'] = df['Torque'].apply(convert_torque)

# Verify the change
print(df['Torque'].dtype)  # Should show float64

# Display the updated DataFrame
print(df)
```
